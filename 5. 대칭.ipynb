# 대칭(Flip, symmetry)


영상이나 이미지를 __대칭__시켜 띄울수 있습니다. __상하__ 또는 __좌우__방향으로 대칭할 수 있습니다.

# Main Code

```python
import cv2
src = cv2.imread("Image\glass_of_water.jpg", cv2.IMREAD_COLOR)
dst = cv2.flip(src, 0)

cv2.imshow("src",src)
cv2.imshow("dst",dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Detailed Code

```python
src = cv2.imread("Image\glass_of_water.jpg", cv2.IMREAD_COLOR)
```

원본 이미지로 사용할 src를 선언하고 이미지를 불러옵니다.

```python
dst = cv2.flip(src,0)
```

결과 이미지로 사용할 dst를 선언하고 대칭 함수를 적용합니다.

cv2.flip(원본 이미지, 대칭방법) 을 의미합니다.

__대칭 방법__은 __상수__를 입력하여 대칭시킬 수 있습니다.

0일 경우, 상하 방향으로 대칭합니다.

1일 경우, 좌우 방향으로 대칭합니다.

```python
cv2.imshow("src",src)
cv2.imshow("dst",dst)
```

cv2.imshow()를 사용하여 이미지를 출력할 수 있습니다.

# Additional Information

대칭 방법 중 __상수__를 0보다 __낮은 값__을 입력할 경우, 상하 대칭으로 간주합니다.

대칭 방법 중 __상수__를 1보다 __높은 값__을 입력할 경우, 좌우 대칭으로 간주합니다.
